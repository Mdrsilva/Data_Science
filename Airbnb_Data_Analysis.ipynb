### **Q5. Are there outliers?** ###

By the histogram distribution, it is possible to verify evidence of the presence of outliers (Those are unusual values, out of line with the pattern of the data set). Outliers present a particular challenge for analysis, and thus it becomes essential to identify, understand and treat these values, as they can give us an incorrect picture of our data.


###_Why outlier exists?_###
There are generally two reasons for the existence of outliers. First, someone may have entered data incorrectly and thus it is an error. These errors can be a result of human error; the system generated or may be a result of some incorrect calculation. Second, there may be real data points which are present because of the nature of the data. There may be data points that result because of skewed data set.

Lets take as example the `price`, `minimum_nights` and `calculated_host_listing_count` variables. The values do not follow a distribution, and they distort the entire graphical representation.

To confirm it, I am going to use two quick ways that help detect outliers. Are they:

* Statistical summary using describe()
* Boxplot plotting and IQR Score Method


```
# Having a general view of the variables
df_airbnb[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe().round(1)
```

Looking at the statistical summary above, we can confirm some hypotheses such as:

The `price` variable has 75% of the values below 262.00 CAD, but its maximum value is 20,000.00 CAD. That is evident of the presence of outliers on this variable. The minimum number of nights (minimum_nigts) is 1 night, but the max value is 399 days (over a year), which shows us there are outliers in the minimum_night column as well.


## Removing Outliers

To address the outliers of this dataset, I am going to check the statistical distribution, plot boxplots and find the upper and the lower limits of the boxplots using the IQR Score.

This method consists in listing the data values in order from least to greatest. When we have the values in ascending order, identify the median (Q2). This value is the midpoint in your data set, which separates the upper 50% from the lower 50%. 

The IQR is calculated subtracting the third quartile (75%) by the first quartile (25%).


# IQR = Q3 - Q1


<p align="center"><img src="https://www.simplypsychology.org/boxplot-outliers.png", width="50%"></p>

Let's take a look at our data and see what we have identified.


```
# Identifying outliers in the price variable
q1_price = df_airbnb.price.quantile(.25)
q3_price = df_airbnb.price.quantile(.75)
IQR_price = q3_price - q1_price
print('Price IQR', IQR_price)

# Defining the upper and the lower limits                                       
sup_price = q3_price + 1.5 * IQR_price
inf_price = q1_price - 1.5 * IQR_price

print('Price upper limit: ', sup_price)
print('Price lower limit: ', inf_price)
```

Here we can see that we do not have lower outlier as the minimum price is zero, I will exclude this from the dataset as it does not make sense has a listing for free. I am also going to exclude the outliers in the upper limit, which represent 6.4% of the data, as we can see below.

#### Boxplot for price



```
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15,3))
df.price.plot(kind='box', vert=False);
ax.set_title('Original Dataset - price')
plt.show()
print("O dataset possui {} colunas".format(df.shape[0]))
print("{} Entradas acima de 358,75".format(len(df[df.price > sup_price])))
print("Representam {:.2f}% do dataset".format((len(df[df.price > sup_price]) / df.shape[0])*100))

```

```
# identificando os outliers para a variável minimum_nights
q1_minimum_nights = df.minimum_nights.quantile(.25)
q3_minimum_nights = df.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('Mininum nights IQR: ', IQR_minimum_nights)

# definindo os limites                                       
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights

print('Limite superior de minimum_nights: ', sup_minimum_nights)
print('Limite inferior de minimum_nights: ', inf_minimum_nights)
```

Aqui podemos ver que, apesar de não termos outliers na parte inferior, as minimum night is 1 night CAD. However, I am going to exclude the outliers in the limite superior, which represent 6.4% of the data, as we can see below.

#### Boxplot para minimum_nights


```
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15,3))
df.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Dataset Original - minimum_nights')
plt.show()
print("O dataset possui {} colunas".format(df.shape[0]))
print("{} Entradas acima de 72 minimum nights".format(len(df[df.minimum_nights > sup_minimum_nights])))
print("Representam {:.2f}% do dataset".format((len(df[df.minimum_nights > sup_minimum_nights]) / df.shape[0])*100))
```

```
# limpando o dataset
df_clean = df.copy()

df_clean.drop(df_clean[df_clean.price > sup_price].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > sup_minimum_nights].index, axis=0, inplace=True)

print('Shape antes da limpeza: ', df.shape)
print('Shape após a limpeza: ',df_clean.shape)
```

```
# verificando o conjunto limpo
fig, ax = plt.subplots(figsize=(15,3))
df_clean.price.plot(kind='box', vert=False);
ax.set_title('Dataset Limpo - price')
plt.show()
print("Tamanho original: {} colunas".format(df.shape[0]))
print("Tamanho atual: {} colunas".format(df_clean.shape[0]))
```

```
# verificando o conjunto limpo
fig, ax = plt.subplots(figsize=(15,3))
df_clean.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Dataset Limpo - minimum nights')
plt.show()
print("Tamanho original: {} colunas".format(df.shape[0]))
print("Tamanho atual: {} colunas".format(df_clean.shape[0]))
```

### **Q4. What is the correlation between the variables**


Correlation means that there is a relationship between two things. We are looking for a relationship or similarity between two variables in our context.

This relationship can be measured, and it is a function of the correlation coefficient to establish its intensity. To identify the existing correlations between the variables of interest, I will:

Create a correlation matrix
Generate a heatmap from this array, using the seaborn library.

```
# Create matrix corelation
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()


# Display matrix corelation
display(corr)
```
